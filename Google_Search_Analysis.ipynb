Then after we import libraries of python which is required for our project.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
```

Now we use build_payload method from Pytrends is used to build a list of keywords you want to search in Google Trends. You can also specify the timeframe to gather data and the category to query the data from.

```python
Trending_topics = TrendReq(hl='en-US', tz=360)
kw_list=["Cloud Computing"] # you can also add a list of keywords
Trending_topics.build_payload(kw_list,cat=0, timeframe='today 12-m')
pytrends = TrendReq(hl='en-US', tz=360)
```

Now we fetch the data and sort it by "Data Science".

```python
pytrends.build_payload(kw_list=["Cloud Computing"])
df = pytrends.interest_by_region()
df.sort_values(by=["Cloud Computing"], ascending=False)
```

Now we can find a trending keywords in particular regions.

We also find that how much particular keywords search in particular region.

```python
pytrends = TrendReq(hl='en-US', tz=360)
pytrends.build_payload(kw_list=['World war'])
df = pytrends.interest_by_region()
df.head(20)
```

We can also find the yearly's graphically pattern of searching that particular keyword.

```python
pytrends = TrendReq(hl='en-US', tz=360)
pytrends.build_payload(kw_list=['Programming'])
pytrends = pytrends.interest_over_time()
fig, ax = plt.subplots(figsize=(10, 5))
pytrends['Programming'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Programming', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(30, 20))
data['Machine Learning'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Machine Learning', fontweight='bold')
plt.xlabel('Year')

plt.ylabel('Total Count')
plt.show()
```
